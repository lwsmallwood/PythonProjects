## Seaborn Visualizations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
plt.rcParams["patch.force_edgecolor"] = True
```

```python
#seaborn has some pre-loaded data sets

df_tips = sns.load_dataset('tips')
```

```python
df_tips.head(3)
```

# Single Variable Distribution

```python
sns.distplot(df_tips['total_bill'])
```

```python
#specify number of bins
sns.distplot(df_tips['total_bill'], kde=False, bins = 30)
```

## Joint Plot - BiVariate Comparions

```python
sns.jointplot(x='total_bill', y='tip', data = df_tips)
```

```python
# can modify what appears in the center with KIND

sns.jointplot(x='total_bill', y='tip', data = df_tips, kind = 'reg')

# other kinds include HEX and KDE
```

```python
#Simple linear model = lm plot

sns.lmplot(x='total_bill', y='tip', data=df_tips)
```

```python
#can make matplotlib modifications

sns.lmplot(x='total_bill', y='tip', data=df_tips, markers='v', scatter_kws={'s':100, 'color':'r'})
```

```python
#seperate by categorical items

sns.lmplot(x='total_bill', y='tip', data=df_tips, col='time', row='sex')
```

## Pair Plot - looking at all variables correlations

```python
sns.pairplot(df_tips)

#Only compares numerical values
```

```python
# add hue argument to seperate categorical elements!
# Palette allows color schemes

sns.pairplot(df_tips, hue='sex', palette = 'plasma')
```

## Categorial Plots

```python
#BARPLOT

sns.barplot(x='sex', y='total_bill', data=df_tips)

# does 'mean average' by default. can pass in ESTIMATOR = NP.STD for standard deviation)
```

```python
#COUNTPLOT

sns.countplot(x='sex', data=df_tips)
```

```python
#BOXPLOTS

sns.boxplot(x="day", y="total_bill", data=df_tips,palette='plasma')

```

```python
#Add another categorical comparison with hue. Change orientation with orient.

sns.boxplot(x="day", y="total_bill", data=df_tips, hue='smoker')
```

```python
#StripPlot

sns.stripplot(x='day', y='total_bill', data=df_tips)
```

```python
#add some noise to more clearly see density with jitter=True, and hue for categorization, and dodge='True' to seperate

sns.stripplot(x='day', y='total_bill', data=df_tips, jitter=True, hue='sex')#, dodge='True')
```

## Heatmaps

```python
tipcorr = df_tips.corr()
```

```python
sns.heatmap(tipcorr, annot=True, cmap='Blues')
```

# Quick Pandas Charts

We can call charts quickly, directly from pandas DFs. Chart types include....

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

```python
df_tipsnum = df_tips.filter(['total_bill', 'tip', 'size'], axis=1)

df_tipsnum['total_bill'].plot.density()
```

```python
df_tipsnum.plot.scatter(x='total_bill',y='tip')
```

```python
df_tipsnum.plot.area()
```
