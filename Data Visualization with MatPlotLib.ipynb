## Basic Functional Plots

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data
x = [ 0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5]
y = [ 0,0.25,1,2.25,4,6.25,9,12.25,16,20.25,25]
```

```python
# simple, functional plot
plt.plot(x, y)
```

```python
#Adding more details
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis T I T L E')
plt.ylabel('Y Axis T I T L E')
plt.title('-- String --')
plt.show()
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

## Object Oriented Plots (more flexible)

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1 - % of canvas)

# Plot on that set of axes
axes.plot(x, y)
```

```python
#Same, but adding more details (Axis, colors etc)

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'm--')
axes.set_xlabel('Set X AXIS Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y AXIS  Label')
axes.set_title('T . I . T . L . E . ')
```

## OO Subplots

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()
```

```python
# Specify rows and colums
fig, axes = plt.subplots(nrows=3, ncols=3)
_
#clean up overlapping text
plt.tight_layout()
```

```python
# Specify rows and colums
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
#axes is just a list. Therefore can iterate through

for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('TITLE')

# Display the figure object    
fig
```

```python
#Specify plot contents with indexing (and change plot size with figsize =)

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,5))
axes[0].plot(x, y, 'r')
axes[1].plot(x, x, 'g')
```

```python
#Simple Change Plot Size

fig, axes = plt.subplots(figsize=(10,3))
axes.plot(x, y, 'm')
```

```python
#Saving image..

fig.savefig("filename.png")
```

## Multi Charts and Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_title("**MAIN TITLE**");

ax.set_xlabel("--x--")
ax.set_ylabel("--y--");

ax.plot(x, x)
ax.plot(x, y)
```

```python
# Plus Legend

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_title("**MAIN TITLE**")

ax.set_xlabel("--x--")
ax.set_ylabel("--y--")

ax.plot(x, x, label='linear')
ax.plot(x, y, label='exponent')


ax.legend()
```

```python
#Markers, colors, etc

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color = '#16a085', lw = 4, ls = '-', marker = 'o', markersize=8)
        
```

```python
#Set axes limits

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, lw=4)

ax.set_ylim([1, 20])
ax.set_xlim([2, 4])

```
